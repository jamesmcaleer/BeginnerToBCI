```python
chans = []
for i in range(8):
    chans.append(f" EXG Channel {i}")

print(df[chans])
```

'iloc' allows me to get specific rows from the DataFrame

```python
df.iloc[0:2]
```

Let's combine this with getting retrieving specific columns to get a chunk of the EEG data I'm looking for...

```python
df[chans].iloc[0:10]
```

Now I'm going to try a couple of options for iterating through this data, sample-by-sample

```python
two_chans = chans[0:2]

for i in range(3):
    print(df[two_chans].iloc[i]) # only showing the first 2 channels and first 3 samples to save space
```

This pre-built function I found 'iterrows()' seems to have the same effect
<br>

I don't seem to have the same control over how many rows I iterate through, and I would imagine iterrows() is faster than me calling iloc() every iteration
<br>

So from here on out, I will be using the pre-built function
<br>

If I don't want to loop through ALL samples, I simply call iloc() once on the dataframe before the loop to chop it

```python
for index, row in df[two_chans].iloc[0:3].iterrows():
    print(f"Sample: {index}", row)
```

Now that I have some control over this data, I want to start looking for peaks, averages, and more
<br>

To start with this, let's explore the 'loc()' function, not to be confused wit 'iloc()'
<br>

To my knowledge, 'loc()' lets me find parts of the DataFrame where a condition is true, similar to the 'WHERE' in SQL
<br>

I'm going to try to show rows where the amplitude is above average in Channel 0
<br>

<hr>

So something weird happened here, when I tried to call 'iterrows()' on df[' EXG Channel 0'], I had an error as I could not call the method on a Series
<br>
Instead, I have to nest the channel in another list, similar to my list 'chans' before so that the type is correct to run the method on

<hr>

Anyways, it was a success. From a programming standpoint, if you are wondering how I used 'index' for calculating my average, the last value of 'index' (and any other incrementing value) retains its final value and can be used outside the loop. Pretty cool

```python
sum_vals = 0
for index, row in df[ [' EXG Channel 0'] ].iterrows():
    sum_vals += row[' EXG Channel 0']

avg = sum_vals / index

df.loc[df[' EXG Channel 0'] > avg][chans]
```
