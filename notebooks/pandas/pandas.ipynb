Now that I have some control over this data, I want to start looking for peaks, averages, and more

To my knowledge, 'loc()' lets me find parts of the DataFrame where a condition is true, similar to the 'WHERE' in SQL

```python
sum_vals = 0
for index, row in df[ [' EXG Channel 0'] ].iterrows():
    sum_vals += row[' EXG Channel 0']

avg = sum_vals / index

df.loc[df[' EXG Channel 0'] > avg][chans]
```

### Bug...
After returning to this a day later, I realize that the code above in the comments will NOT work
<br>
Let's get into it:
<br>

Since the 'loc()' method takes in conditions, I thought I would be able to AND the 8 channels WHERE their values were greater than the average.
<br>
This is why you see me looping through and using '&='. However this will not work because what we are trying to do with 'loc()' is filter the DataFrame
<br>
where certain conditions are true. Pandas has its _own_ syntax for ANDs, instead of 'and' it is '&'.
<br>

The reason for this is because the ANDs we are aiming to do don't actually fit the way an 'and' works in Python. Instead, 'loc()' narrows down the df
<br>
from the first condition, then uses its own '&' to see we should futher narrow down the df to the next condition. My above code simply creates a 
<br>
boolean variable in which we AND the current channel being greater than the average to the variable. This doesn't actually locate the rows WHERE
<br>
the values are greater.
<br>

I though I'd type out my entire thought process here to make sure I really understood the bug. In the future of this blog, I might only highlight the most
<br>
important bugs in a manner like this.

### Fixed
Sadly, until I can think of a better solution, we will have to brute force the custom '&'s that Pandas uses

## Normalizing EEG Channels
To wrap up this blog, I want to challenge myself to normalize all of the values in the 8 channels of this DataFrame

So after some research, the formula for the normalization I want to do (min-max normalization) is, for each element of a column, 
<br>
(x_i - x_min) / (x_max - m_min). In this case, x_i is the ith element of the column.
<br>

To do this normalization there are a couple of steps I need to take:
- Find the min and max of the column
- On each iteration, locate the element at position i
- Apply the formula and put this new value in the _result_ DataFrame
- Repeat for all 8 channels

```python
# df[chans].max().iloc[0] | we will be using this logic from this line created earlier in the blog
maxs = []
mins = []

df_max = df[chans].max() # make it into its own variable to save from finding the max every iteration
df_min = df[chans].min()

result = {} # this will soon be turned into a DataFrame
for i in range(8):
    maxs.append(df_max.iloc[i])
    mins.append(df_min.iloc[i])
    result[f'Channel {i}'] = []

normalized = pd.DataFrame(result)

for index, row in df[chans].iterrows(): 
    for i in range(8):
        x_i = row[f' EXG Channel {i}']
        normalized.loc[index, f'Channel {i}'] = (x_i - mins[i]) / (maxs[i] - mins[i]) 
        # so if i want to change the value of an element in row x and column y, i should use '.loc[x, y]'

normalized.head()
```

Great! We have a new DataFrame of the normalized values.
